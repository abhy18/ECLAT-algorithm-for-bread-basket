```python
grped = df.groupby('Transaction')['Item'].apply(list).reset_index()
expand = grped['Item'].apply(pd.Series)
expand
```

```python
from pyECLAT import ECLAT
ec = ECLAT(data=expand, verbose=True)
```

```python
ec.uniq_
```

```python
indexs , support = ec.fit(min_support=.02,min_combination=1,max_combination=3,separator = ' & ',verbose = True)
```
